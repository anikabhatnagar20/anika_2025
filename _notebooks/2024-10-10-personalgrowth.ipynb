```java
// Abstract class Animal, the base for all animals
abstract class Animal {
    String name;

    // constructor to set the name
    Animal(String name) {
        this.name = name;
    }

    // abstract method for animal sound
    abstract String sound();

    // abstract method to print a unique behavior
    abstract void uniqueBehavior();
}

// Dog class that extends Animal
class Dog extends Animal {
    String breed;

    // constructor for Dog
    Dog(String name, String breed) {
        super(name);
        this.breed = breed;
    }

    // Dog's version of sound
    @Override
    String sound() {
        return "Barks";
    }

    // unique behavior for Dog
    @Override
    void uniqueBehavior() {
        System.out.println("This is a " + breed + " dog, and it loves to fetch balls!");
    }
}

// Cat class that extends Animal
class Cat extends Animal {
    String color;

    // constructor for Cat
    Cat(String name, String color) {
        super(name);
        this.color = color;
    }

    // Cat's version of sound
    @Override
    String sound() {
        return "Meows";
    }

    // unique behavior for Cat
    @Override
    void uniqueBehavior() {
        System.out.println("This is a " + color + " cat, and it loves to climb trees!");
    }
}

// Bird class that extends Animal
class Bird extends Animal {
    double wingspan;

    // constructor for Bird
    Bird(String name, double wingspan) {
        super(name);
        this.wingspan = wingspan;
    }

    // Bird's version of sound
    @Override
    String sound() {
        return "Chirps";
    }

    // unique behavior for Bird
    @Override
    void uniqueBehavior() {
        System.out.println("This bird has a wingspan of " + wingspan + " meters, and it loves to soar in the sky!");
    }
}

// Main class to run the code
public class AnimalTest {
    public static void main(String[] args) {
        // creating instances of the animals
        Dog dog = new Dog("Buddy", "Golden Retriever");
        Cat cat = new Cat("Whiskers", "Black");
        Bird bird = new Bird("Sky", 1.5);

        // print sound and unique behavior for each
        System.out.println(dog.name + " " + dog.sound());
        dog.uniqueBehavior();

        System.out.println(cat.name + " " + cat.sound());
        cat.uniqueBehavior();

        System.out.println(bird.name + " " + bird.sound());
        bird.uniqueBehavior();
    }
}

AnimalTest.main(null);

```
